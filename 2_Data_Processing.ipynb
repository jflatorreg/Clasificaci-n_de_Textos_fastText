```python
import pandas as pd
import os
import re
import unidecode 
import difflib
```

```python
files = os.listdir('./data/')
```

```python
df = []
for i, file in enumerate(files):
    print('Processing file:', i+1, '... '+ str(len(files)))
    df_tmp = pd.read_csv('./data/' + file)
    df.append(df_tmp)
```

# Clean Data

```python
df.drop_duplicates(subset=['tweet_id', 'text'], keep='last', inplace=True, ignore_index=True)
len(df)
```

```python
df['txt'] = df['text'].str.lower()
df['txt'] = df['txt'].str.replace('co','')
df['txt'] = df['txt'].str.replace('ciclo ruta','ciclorruta')
```

```python
def clean_txt(txt):  
    clean = ' '.join(re.findall('[^!,.Â¿? ]+', txt))
    clean = re.sub(r'rt', '', clean)
    clean = re.sub(r'@\S+', '', clean)
    clean = re.sub(r'#\S+', '', clean)
    clean = re.sub(r"http\S+", "", clean)
    clean = re.sub(r'/\S+', '', clean)
    clean = re.sub(r'\W+', ' ', clean)
    clean = unidecode.unidecode(clean)
    clean = clean.lstrip()
    clean = clean.rstrip()
    return clean
```
